# Dataset Statistics
---

```python
import os
import pandas as pd
import numpy as np
import pretty_midi as pm

statistics = pd.DataFrame(columns=['Composer', 'N_midis', 'Min_duration', 'Max_duration', 'Ave_duration', 'Duration'])

N_composers = len(os.listdir('midis'))
for i, composer in enumerate(os.listdir('midis')):
    print(i+1, '/', N_composers, end='\r')
    MIDIs = os.listdir(os.path.join('midis', composer))
    
    durations = []
    for midi in MIDIs:
        midi_data = pm.PrettyMIDI(os.path.join('midis', composer, midi))
        duration = midi_data.get_end_time()
        durations.append(duration)
        
    statistics.loc[i] = [composer, len(MIDIs), np.min(durations), np.max(durations), np.mean(durations), np.sum(durations)]
    
print()
Total_row = dict(Composer='Totol',
                N_midis=np.sum(statistics.loc[:,'N_midis']),
                Min_duration='--',
                Max_duration='--',
                Ave_duration='--',
                Duration=np.sum(statistics.loc[:,'Duration']))
statistics = statistics.append(Total_row, ignore_index=True)
display(statistics)
```
